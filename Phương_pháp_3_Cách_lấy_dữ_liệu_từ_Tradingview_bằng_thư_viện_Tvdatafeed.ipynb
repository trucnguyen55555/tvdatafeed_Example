# **Phần II. Đăng nhập tài khoản Tradingview**

```python
from tvDatafeed import TvDatafeed, Interval

username = 'your username'
password = 'your password'

tv = TvDatafeed(username, password)
```

# **Phần III. Download dữ liệu lịch sử**

```python
# Dữ liệu lịch sử của mã chỉ số
vndata = tv.get_hist(symbol='VNINDEX',exchange='HOSE',interval=Interval.in_1_hour,n_bars=1000)
vndata.head()
```

```python
# Dữ liệu lịch sử của mã chứng khoán
acbdata = tv.get_hist(symbol='ACB',exchange='HOSE',interval=Interval.in_1_hour,n_bars=1000)
acbdata.head()
```

```python
# Dữ liệu lịch sử của chỉ số hàng hóa
commodity_data = tv.get_hist(symbol='CL1!', exchange='NYMEX', interval=Interval.in_1_hour, n_bars=1000)
commodity_data.head()
```

```python
# Dữ liệu lịch sử của Crypto
btc_data = tv.get_hist(symbol='BTCUSD', exchange='COINBASE', interval=Interval.in_1_hour, n_bars=1000)
btc_data.head()
```

# **Phần III. Trực quan hóa dữ liệu**

```python
# Nhập các thư viện trực quan hóa dữ liệu
import plotly.subplots as ms
import plotly.graph_objects as go

# Tạo subplot với 2 hàng
fig = ms.make_subplots(rows=2, cols=1, shared_xaxes=True, vertical_spacing=0.02)

# Thêm biểu đồ Candlestick cho dữ liệu 'ACB'
fig.add_trace(go.Candlestick(x=acbdata.index,
                             open=acbdata['open'],
                             high=acbdata['high'],
                             low=acbdata['low'],
                             close=acbdata['close'],
                             name='Candlestick'),
              row=1, col=1)

# Thêm biểu đồ Volume dưới dạng cột
fig.add_trace(go.Bar(x=acbdata.index, y=acbdata['volume'], name='Volume'),
              row=2, col=1)

# Cập nhật bố cục của biểu đồ
fig.update_layout(title='ACB Candlestick Chart',
                  yaxis_title='Price',
                  xaxis_title='Date',
                  xaxis_rangeslider_visible=False,
                  bargap=0.3,)  # Khoảng cách giữa các cột Volume

# Hiển thị biểu đồ
fig.show()
```
